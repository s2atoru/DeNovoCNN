## モデルの読み込み

```python
snp_model = "models/snp"
del_model = "models/del"
ins_model = "models/ins"


models_cfg = {
    "snp_model": snp_model,
    "del_model": del_model,
    "ins_model": ins_model,
}

models_dict = {
    "Substitution": tf.keras.models.load_model(models_cfg["snp_model"], compile=False),
    "Deletion": tf.keras.models.load_model(models_cfg["del_model"], compile=False),
    "Insertion": tf.keras.models.load_model(models_cfg["ins_model"], compile=False),
}
```

```python
model = tf.keras.models.load_model("models/del.h5")
```

```python
# model = models_dict["Deletion"]
model.summary()
```

```python
IMAGE_DATA_DIR = "data/publish_images"

train_dir = os.path.join(IMAGE_DATA_DIR, "deletion", "train")

dnm_dir = os.path.join(train_dir, "DNM")
iv_dir = os.path.join(train_dir, "IV")
```

```python
# import png images from dnm_dir and save as numpy array
dnm_images = load_images_from_directory(dnm_dir)
```

```python
# show dnm_images[0]

plt.imshow(dnm_images[0])
plt.axis("off")
plt.show()
```

```python
# show the size of dnm_images[0]
print("Size of dnm_images[0]:", dnm_images[0].shape)
```

```python
# show the max and min values in dnm_images[0]
print("Max value in dnm_images[0]:", np.max(dnm_images[0]))
print("Min value in dnm_images[0]:", np.min(dnm_images[0]))
```

```python
# show the number of images in dnm_images
print(f"Number of images: {len(dnm_images)}")
```

```python
dnm_images_array_normalized = preprocess_images(dnm_images)
print(f"dtype of dnm_images_array_normalized: {dnm_images_array_normalized.dtype}")
print(
    f"show the shape of dnm_images_array_normalized: {dnm_images_array_normalized.shape}"
)
```

```python
# show the max and min values in dnm_images_array_normalized[0]
print(
    "Max value in dnm_images_array_normalized[0]:",
    np.max(dnm_images_array_normalized[0]),
)
print(
    "Min value in dnm_images_array_normalized[0]:",
    np.min(dnm_images_array_normalized[0]),
)
```

```python
# make predictions using model for dnm_images_array
predictions = model.predict(dnm_images_array_normalized)
```

```python
predictions.shape
```

```python
# show the distribution of predictions
plt.hist(predictions, bins=30)
plt.title("Distribution of Predictions")
plt.xlabel("Prediction Value")
plt.ylabel("Frequency")
plt.show()
```
